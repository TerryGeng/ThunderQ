```python
import numpy as np
import matplotlib.pyplot as plt

from waveform import *

def plot(wave):
    raw = wave.sample(50)
    plt.plot(raw.sample_points, raw.data)
    plt.show()
    
    
def plot_complex(wave):
    raw = wave.sample(50)
    plt.plot(raw.sample_points, raw.data.real, color='b')
    plt.plot(raw.sample_points, raw.data.imag, color='r')
    plt.show()
```

```python
sin_wave = Sin(10, 1, 2, 0)
cos_wave = Cos(10, 1, 2, 0) * 1j

sum_wave = SumWave(sin_wave, cos_wave)

raw_waveform = sum_wave.sample(10)

plt.plot(raw_waveform.sample_points, raw_waveform.data.real)
plt.plot(raw_waveform.sample_points, raw_waveform.data.imag, color='r')
plt.show()
```

```python
exp_wave = ComplexExp(10, 1, 2, 0)

raw_waveform = exp_wave.sample(10)

plt.plot(raw_waveform.sample_points, raw_waveform.data.real)
plt.plot(raw_waveform.sample_points, raw_waveform.data.imag, color='r')
plt.show()
```

```python
blank_wave = Blank(5)
sin_wave = Sin(10, 1, np.pi, 0)
gaussian_env = Gaussian(10, 1)
seq_wave = blank_wave.concat(gaussian_env * sin_wave).concat(blank_wave)

raw = seq_wave.sample(10)

plt.plot(raw.sample_points, raw.data)
plt.show()
```

```python
iq_wave = CalibratedIQ(5, seq_wave, Blank(20))
plot_complex(iq_wave)
```
